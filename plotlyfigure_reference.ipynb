<font size=5>OVERLOADED MATLAB PLOTTING FUNCTIONS</font>

Native MATLAB graphing functions have been included as overloaded methods of <code>plotlyfigure.m </code>. These methods behave excatly as their corresponding MATLAB counterparts with the additional functionality of automatically populating the <a>data</a> and <a>layout</a> properties of the plotlyfigure.m object and making a direct call to Plotly via the <code>plotly</code> method. These methods can be invoked by passing a plotlyfigure object as the first argument to the native MATLAB plotting function. 

<b>EXAMPLE:</b>   

<code>>> p = plotlyfigure.m;</code><br>
<code>>> plot(p, ... );</code><br>

# REFERENCE

<br>
<font size=6>PROPERTIES</font>
<hr>

<code><b id="data">data</b> [cell array]</code>

>The <a href="https://plot.ly/matlab/reference/#Trace_objects">data traces</a> associated with the Plotly figure.

<code><b id="layout">layout</b> [structure array]</code>

> The <a href="https://plot.ly/matlab/reference/#layout">layout structure</a> associated with the Plotly figure.

<code><b id="PlotOptions">PlotOptions</b> [structure array]</code>

> The Plot options associated with the creation of the <code>plotlyfigure.m</code> object within MATLAB and the online figure created on your Plotly account. The following fields are contained withthin the PlotOptions property: 

><code><b>FileName</b> [string]:</code>

>The name of the figure as it will appear at the top of the figure in MATLAB and the filename associated with the figure created on Plotly. By default <code>Filename = 'myplotlyfigure'</code> if no filename is provided upon the initialization of the plotlfigure object. <br>

><code><b>FileOpt</b> [string]:</code> 

> The <a href="https://plot.ly/matlab/file-options/">file write option</a> specifying how each new call to Plotly will affect the plot in your workspace of the same name. By default <code>FileOpt = 'overwrite'</code>. Set <code>FileOpt = 'new' or 'append' or 'extend'</code> as desired. <br>

><code><b>WorldReadable</b> [boolean]:</code>

>The <a href="https://plot.ly/matlab/privacy/">privacy of the figure</a> once sent to Plotly. By default, <code> WorldReadable = true </code> for public graphs. For private graphs, set <code> WorldReadable = false </code>.<br>

><code><b>OpenURL</b> [boolean]:</code>

>A boolean switch used for displaying/suppressing the figure hyperlink when a call is made to Plotly. By default, <code> OpenUrl = true </code> and the URL is displayed (if a MATLAB command window is present). Set <code> OpenUrl = false </code> to suppress displaying the figure hyperlink.<br>

><code><b>Visible</b> [boolean]:</code> 

>A boolean switch used for displaying/supressing the MATLAB figure upon creation of the plotlyfigure object. By default <code>Visible = true</code> and the MATLAB figure is displayed. Set <code>Visible = false</code> to prevent the MATLAB figure from being displayed upon creation. <br>

<code><b id="State">State</b> [structure array]<br></code>

>The state property houses infromation regarding the MATLAB figure, axes, plots, texts, colorbars, and legends involved in the creation of the Plotly figure. The fields contained with the <code>State</code> property structure the information on each of these MATLAB graphics objects as follows: 
å
><code><b>Figure</b> [structure array]:</code>

> The <code>Figure</code> field of the <code>State</code> property is a structure array with the following fields:

><i>Handle</i>: graphics handle associated with the figure.<br>
><i>NumAxes</i>: number of axes parented by the figure.<br>
><i>NumPlots</i>: number of plots parented by the figure.<br>
><i>NumLegends</i>: number of legends parented by the figure.<br>
><i>NumColorbars</i>: number of colorbars parented by the figure.<br>
><i>NumTexts</i>: number of text parented by the figure.<br>

><code><b>Axis</b> [array of structs]:</code>

> The <code>Axis</code> field of the <code>State</code> property is a structure array with the following fields:

><i>Handle</i>: graphics handle associated with the axis.

><code><b>Plot</b> [array of structs]:</code>

> The <code>Plot</code> field of the <code>State</code> property is a structure array with the following fields:

><i>Handle</i>: graphics handle associated with the plot.<br>
><i>AssociatedAxis</i>: graphics handle of the axis parenting the plot.<br>
><i>Class</i>: name of graphing class associated with the plot.<br>

><code><b>Text</b> [array of structs]:</code>

> The <code>Text</code> field of the <code>State</code> property is a structure array with the following fields:

><i>Handle</i>: graphics handle associated with the plot.<br>
><i>AssociatedAxis</i>: graphics handle of the axis parenting the text.<br>
><i>Title</i>: boolean indicating if the text is an axis title (true).<br>

><code><b>Colorbar</b> [array of structs]:</code>

> The <code>Colorbar</code> field of the <code>State</code> property is a structure array with the following fields:

><i>Handle</i>: graphics handle associated with the colorbar.<br>
><i>AssociatedAxis</i>: graphics handle of the axis peered with the colorbar.<br>
    
><code><b>Legend</b> [array of structs]:</code>

> <i>Handle</i>: graphics handle associated with the legend.<br>
><i>AssociatedAxis</i>: graphics handle of the axis peered with the legend.<br>

<b id="PlotlyDefaults"><code>PlotlyDefaults</code></b> [structure array] 

>The defaults used to smooth the transition between the MATLAB and Plotly figures. These settings can be changed to suit your personal preference. The current Plotly Defaults are set as follows:
å
> <code>PlotlyDefaults.MinTitleMargin = 80; </code> <br>
> <code>PlotlyDefaults.TitleHeight = 0.01; </code> <br>
> <code>PlotlyDefaults.FigureIncreaseFactor = 1.5; </code> <br>
> <code>PlotlyDefaults.AxisLineIncreaseFactor = 1.5; </code> <br>
> <code>PlotlyDefaults.MarginPad = 0; </code> <br>
> <code>PlotlyDefaults.MaxTickLength = 20; </code> <br>
> <code>PlotlyDefaults.ExponentFormat = 'none'; </code> <br>
> <code>PlotlyDefaults.ErrorbarWidth = 6; </code> <br>
> <code>PlotlyDefaults.ShowBaselineLegend = false; </code> <br>
> <code>PlotlyDefaults.Bargap = 0;</code> <br>

<code><b id="UserData">UserData</b> [structure array]</code>

>The data associated with the user and the user experience. The fields contained within the UserData structure are as follows: å

><code><b>Username</b> [string]:</code> 

>The username associated with your <a href="https://plot.ly/how-to-sign-up-to-plotly/">Plotly account</a>. 

><code><b>ApiKey</b> [string]:</code>

>The API key associated with your <a href="https://plot.ly/how-to-sign-up-to-plotly/">Plotly account</a>. 

><code><b>PlotlyDomain</b> [string]:</code>

>The endpoint url for the REST API. By default the Plotly domain is set to https://plot.ly . This may change depening on your Plotly subscription.

><code><b>Verbose:</b> [boolean]:</code> 

>A boolean switch used to enable/disable the output of the <code>plotlyfigure</code> object's <a href="#validate"><code>validation<code></a> method. By default, <code>Verbose = true</code> and any errors found from the data/layout keys validation will be outputted. Set <code>Verbose = false</code> to supress this output. <br>

<code><b id="Response">Response</b> [structure array]</code>

>A structure array containing the response returned once successfully making a call to the Plotly REST API. The <code>Response</code> property is only populated once the <a href="plotly"><code>plotly</code></a> method is invoked. The fields of the <code>Response</code> property are as follows:
å
><code><b>filename</b> [string]:</code>

>The filename of the plot as saved in your user account.<br>

><code><b>url</b> [string]:</code>

>A url where you can view the rendered graph of your data.<br>

><code><b>error</b>: [string]:</code>

>A string describing a possible error that may have occured. Example: 
<code>Error: A user with this username already exists</code><br>

><code><b>warning</b> [string]:</code>

>A string describing a possible warning message. A string describing a possible warning message. Example: <code>Warning: You are using an outdated version of the API</code><br>

><code><b>message</b> [string]:</code> 

>A string describing a non-critical message.
Example: <code>Message: Plotly has recently expanded the API library.</code><br>

<b id="pf"><code>plotlyfigure</code></b>: Class constructor<br>

>**<code>[INPUT]:</code>**

><code> >> p = plotlyfigure;</code><br>
><code> >> p = plotlyfigure(**fig_han**)</code><br>
><code> >> p = plotlyfigure(**fig_han**, ...,'**Property**','*Value*',...)</code><br>
><code> >> p = plotlyfigure(...,'**Property**','*Value*',...)</code><br>

><hr>

>**<code>[WHERE]:</code>**

><code>**fig_han**: [handle]</code>

>*Figure handle to be parsed*

>**<code>[PROPERTIES / VALUES]:</code>**

><code>**filename**: [string]{'plotlyfigure'}</code>

>*Name of the Plotly figure*

><code>**fileopt**: [string]{'overwrite'}</code> 

>*new, overwrite, extend, append*

><code>**world_readable**[boolean]{true}</code>

>*true = public , false = private*

><code>**open**[boolean]{true}</code>

>*true = show hyperlink (if command window), false = no hyperlink*

><code>**visible**[boolean]{true}</code>

>*true = show MATLAB figure, false = hide MATLAB figure*

><code>**data**[cell array]{}</code>

>*Initial data cell array* 

><code>**layout**[structure array]{struct()}</code>

>*Initial layout structure array*

><hr>

>**<code>[OUTPUT]:</code>**

><code>**p** [object]</code>

>*plotlyfigure object*
